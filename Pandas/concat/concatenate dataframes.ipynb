```python
import pandas as pd
```

```python
#creating my own dataFrames
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york", "chicago", "orland"],
    "temperature": [21, 14, 35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
df = pd.concat([india_weather, us_weather], ignore_index = True)
df
```

```python
df = pd.concat([india_weather, us_weather], keys = ['india', 'us'])
df
```

```python
#creating new data frames to study columns wise concatenation

india_temperature = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [21, 14, 35],
}, index = [0, 1, 2])
india_temperature
```

```python
#give 
india_humidity = pd.DataFrame({
    "city": ["banglore", "delhi", "mumbai"],
    "humidity": [80, 60, 78],
}, index = [2, 1, 0])
india_humidity
```

```python
df = pd.concat([india_temperature, india_humidity])
df
```

```python
df = pd.concat([india_temperature, india_humidity], axis = 1)
df
```

```python
#add series to a dataframe

#create a series
s = pd.Series(["Humid", "Dry", "Rain"], name = "event")
```

```python
df = pd.concat([india_temperature, s], axis = 1)
df
```
