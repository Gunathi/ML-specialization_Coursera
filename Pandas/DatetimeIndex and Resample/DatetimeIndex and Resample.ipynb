```python
import pandas as pd
df = pd.read_csv("aapl.csv", parse_dates = ['Date'], index_col = 'Date')
df.head()
```

### Resample

```python
df.Close.resample('M').mean()
#this gives the mean close prices of each month
#we can use any aggregate function
#There are many resampling frequencies
    # 'D' => daily
    # 'W' => weekly
    # 'M' => monthly
    # 'Q' => quarterly
    # 'A' => annual
    # 'AS' => start of the year
```
