```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([300.0, 500.0, 600.0])
x, y
```

```python
m = x.shape[0]
m
```

```python
plt.scatter(x, y, color='red', marker='*')
plt.title("House pricing")
plt.ylabel("Price (USD)")
plt.xlabel("Size (sqrt)")
plt.show()
```

```python
w = 175
b = 100
```

```python
def compute_output(x, w, b, m):
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_output(x, w, b, m)

plt.plot(x, tmp_f_wb, color='blue', label="Prediction")
plt.scatter(x, y, color='red', marker='*', label='Actual data')

plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
